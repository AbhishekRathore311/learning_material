```python
import numpy as np
```

```python
pop_dict = {
    "conway": 9,
    "dhoni": 7.5,
    "saha": 7.5,
    "gaikwad": 9,
    "gill": 9,
    "dube": 8.5,
    "miller": 8.5,
    "rahane": 8,
    "shankar": 7.5,
    "sudharsan": 7.5,
    "rayudu": 6.5,
    "jadeja": 9,
    "pandya": 9,
    "ali": 7.5,
    "shami": 9,
    "rashid": 9,
    "chahar": 8.5,
    "mohit": 8.5,
    "tushar": 8,
    "pathirana": 8,
    "noor": 8,
    "theekshana": 7
}
pop_list = list(pop_dict.keys())
pop_prob = list(pop_dict.values())
pop_prob = [p/sum(pop_prob) for p in pop_prob]
print(pop_list)
print(pop_prob)
```

```python
for i in range(18):
    print(f"Team {i + 1}")
    team = list(np.random.choice(pop_list, size=11, replace=False, p=pop_prob))
    team_dict = {k: v for k, v in pop_dict.items() if k in team}
    team_list = list(team_dict.keys())
    team_prob = list(team_dict.values())
    team_prob = [p/sum(team_prob) for p in team_prob]
    c_vc = list(np.random.choice(team_list, size=2, replace=False, p=team_prob))
    print("Team")
    print(team)
    print("c_vc")
    print(c_vc)
    print()
```
